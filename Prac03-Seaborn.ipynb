```python
from google.colab import drive
import matplotlib.pyplot as plt
import random
import pandas as pd
import numpy as np
import seaborn as sns
import os
```

```python
os.chdir(APP_ROOTD)
cars_data = pd.read_csv('Toyota.csv', index_col = 0, na_values = ['??', '???'])
cars_data.size
```

```python
cars_data.dropna(axis = 0, inplace = True)
cars_data.size
```

```python
cars_data.head()
```

```python
sns.set(style = 'darkgrid')
sns.regplot(x = cars_data['Age'], y = cars_data['Price'])
```

```python
# without line use fit_reg = False
sns.regplot(x = cars_data['Age'], y = cars_data['Price'], fit_reg = False)
```

```python
# without line use fit_reg = False
sns.regplot(x = cars_data['Age'], y = cars_data['Price'], fit_reg = False, marker = '+')
```

```python
sns.lmplot(x = 'Age', y = 'Price', data = cars_data, fit_reg = False, hue = "FuelType", palette = 'Set1')
```

```python
sns.distplot(cars_data['Age']) # "kde = False" will not show density removes the density line
```

```python
sns.histplot(cars_data['Age'])
```

```python
sns.distplot(cars_data['Age'], bins = 2)
```

```python
sns.countplot(x = "FuelType", data = cars_data)
```

```python
sns.countplot(x = "FuelType", data = cars_data, hue = 'Automatic')
```

```python
pd.crosstab(index = cars_data['Automatic'], columns = cars_data['FuelType'], dropna = True)
```

```python
sns.boxplot(x = cars_data['FuelType'], y = cars_data['Price'])
```

```python
sns.boxplot(x = cars_data['FuelType'], y = cars_data['Price'], hue = cars_data['Automatic'])
```

```python
f, (ax_box, ax_hist) = plt.subplots(2, gridspec_kw={'height_ratios': [0.25, 0.75]})
sns.boxplot(x = cars_data['Price'], ax=ax_box)
sns.histplot(x = cars_data['Price'], ax=ax_hist, bins = 20)
```

```python
sns.pairplot(cars_data, kind='scatter', hue='FuelType', diag_kws={'bw': 0.1})
plt.show()
```
