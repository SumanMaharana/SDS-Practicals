```python
import random
number_range = 1000000
data = pd.DataFrame({
    'A' : [random.choice(('a', 'b' ,'c')) for i in range(number_range)],
    'B' : [random.randint(1, 100) for i in range(number_range)],
    'C' : [random.randint(1, 100) for i in range(number_range)]
})
data
```

```python
mean_A = data.groupby('A').mean()
mean_A
```

```python
mean_A = data.groupby('A')['B'].mean().rename('D').reset_index()
mean_A
```

```python
merge_data = data.merge(mean_A)
```
